# MARCUS merged-lidar radar product

```python
# Load the merged lidar-radar product
current_file = lidar_radar_merged_list[3]
lidar_radar_merged = xr.open_dataset(f'marcus_merged_lidar_radar/{current_file}', decode_times = False)
date_str = current_file.split('_')[-1].split('.')[1][-8:]

# Convert the 'height' variable to metres
lidar_radar_merged['height'] = lidar_radar_merged['height'] * 1000

# Convert lidar_radar_merged 'hour' coordinate to an actual datetime called 'time'
# Convert date_str to datetime object
date_obj = datetime.datetime.strptime(date_str, '%Y%m%d')
# Convert hour coordinate to datetime
minutes_since_start_of_day = np.round(lidar_radar_merged['hour'] * 60)  # Convert hours to seconds
# Create a timedelta array using minutes_since_start_of_day
timedelta_array = np.array([datetime.timedelta(minutes=int(minutes)) for minutes in minutes_since_start_of_day])
# Convert timedelta array to datetime array by adding to the current date
datetime_array = date_obj + timedelta_array
# Assign values of datetime_array to the 'hour' coordinate in test
lidar_radar_merged['hour'] = datetime_array
# Rename the 'hour' coordinate to 'time'
lidar_radar_merged = lidar_radar_merged.rename({'hour': 'time'})

# Additionally, create a total backscatter variable and add it to the dataset
lidar_radar_merged['merged_backscatter'] = np.log10(10**lidar_radar_merged['copol_backscatter'] + 10**lidar_radar_merged['crosspol_backscatter'])

lidar_radar_merged
```

```python
# Create regridded data
regridded_lidar_radar = np.full((len(um_reg_times),len(um_reg_heights)),np.nan)

# Set current loop values
i, j = 43, 40

# Select the target data array in the obs
lidar_radar_merged_slice = lidar_radar_merged.sel(height = slice(um_reg_heights[j], um_reg_heights[j + 1]), time=slice(um_reg_times[i], um_reg_times[i + 1]))

# Keep only cld_phase
variables_to_keep = ['time', 'height', 'cld_phase']
lidar_radar_merged_slice = lidar_radar_merged_slice.drop_vars([var for var in lidar_radar_merged_slice.variables if var not in variables_to_keep])
# print(um_reg_heights[j], um_reg_heights[j + 1])
# print(lidar_radar_merged_slice.height.values)
# print(um_reg_times[i], um_reg_times[i + 1])
# print(lidar_radar_merged_slice.time.values)

# Convert xarray Dataset variable to a DataArray
cld_phase = lidar_radar_merged_slice['cld_phase']

# Extract values from DataArray
cld_phase_values = cld_phase.values
# Modify cld_phase_values according to the following rules
cld_phase_values[cld_phase_values == 8] = 0 # Clear sky
cld_phase_values[np.isin(cld_phase_values, [3, 7])] = 1 # Liquid
cld_phase_values[np.isin(cld_phase_values, [5, 6])] = 2 # Ice
cld_phase_values[cld_phase_values == 4] = 3 # Mixed phase
# print(cld_phase_values)

# Use np.unique on the values array
unique_elements, indices = np.unique(cld_phase_values, return_index=True)
print('unique_elements:', unique_elements)
print(indices)
print(np.arange(len(indices)))

# 0=Clear; 1=Liquid Warm; 2=Ice Virga; 3=Supercooled Liquid; 4=mixed phase; 5=Ice or Mixed phase; 6=Ice; 7=Liquid Virga; 8=Aerosol
# identify blank
if 0 in unique_elements:
    indices[0] = np.count_nonzero(cld_phase_values == 0)
for k in np.arange(len(indices)):
    if indices[k]/sum(indices) > 0.9:
        regridded_lidar_radar[i, j] = unique_elements[k]
        print('UM grid cell set to common value')
    elif 0.1 < indices[np.where(unique_elements == 2)]/sum(indices) < 0.9:
        regridded_lidar_radar[i, j] = 3 # Mixed-phase
        print('UM grid cell set to mixed phase')
    elif (indices[np.where(unique_elements == 2)]/sum(indices) <= 0.1) & (indices[np.where(unique_elements == 3)]/sum(indices) >= 0.5):
        regridded_lidar_radar[i, j] = 3 # Mixed-phase
        print('UM grid cell set to ice')
    else:
        regridded_lidar_radar[i, j] = 1 # Liquid
        print('UM grid cell set to liquid')

non_nan_values = regridded_lidar_radar[~np.isnan(regridded_lidar_radar)]
print(non_nan_values)
```

```python
# Create regridded data
regridded_lidar_radar = np.full((len(um_reg_times),len(um_reg_heights)),np.nan)
variables_to_keep = ['time', 'height', 'cld_phase']

# Loop across time timension
for i in np.arange(len(um_reg_times) - 1):
    # Loop across height dimension
    for j in np.arange(len(um_reg_heights) - 1):
        # Select the target data array in the observations
        lidar_radar_merged_slice = lidar_radar_merged.sel(height = slice(um_reg_heights[j], um_reg_heights[j + 1]), time=slice(um_reg_times[i], um_reg_times[i + 1]))
        # Keep only cld_phase
        lidar_radar_merged_slice = lidar_radar_merged_slice.drop_vars([var for var in lidar_radar_merged_slice.variables if var not in variables_to_keep])
        cld_phase_values = lidar_radar_merged_slice['cld_phase'].values
        # Modify cld_phase_values according to the following rules
        cld_phase_values[cld_phase_values == 8] = 0 # Clear sky
        cld_phase_values[np.isin(cld_phase_values, [3, 7])] = 1 # Liquid
        cld_phase_values[np.isin(cld_phase_values, [5, 6])] = 2 # Ice
        cld_phase_values[cld_phase_values == 4] = 3 # Mixed phase
        unique_elements, indices = np.unique(cld_phase_values, return_index=True)
        # Identify blank
        # 0=Clear; 1=Liquid Warm; 2=Ice Virga; 3=Supercooled Liquid; 4=mixed phase; 5=Ice or Mixed phase; 6=Ice; 7=Liquid Virga; 8=Aerosol
        if 0 in unique_elements:
            indices[0] = np.count_nonzero(cld_phase_values == 0)
        for k in np.arange(len(indices)):
            if indices[k]/sum(indices) > 0.9:
                regridded_lidar_radar[i,j] = unique_elements[k]
            elif 0.1 < indices[np.where(unique_elements == 2)]/sum(indices) < 0.9:
                regridded_lidar_radar[i,j] = 3 # Mixed-phase
            elif (indices[np.where(unique_elements == 2)]/sum(indices) <= 0.1) & (indices[np.where(unique_elements == 3)]/sum(indices) >= 0.5):
                regridded_lidar_radar[i,j] = 3 # Mixed-phase
            else:
                regridded_lidar_radar[i,j] = 1 # Liquid
```

```python
cld_phase = lidar_radar_merged['cld_phase']

# Create a new variable cloud_phase_mod to match the final cloud phase types of Desai et al. (2023)
# 0=Clear; 1=Liquid Warm; 2=Ice Virga; 3=Supercooled Liquid; 4=mixed phase; 5=Ice or Mixed phase; 6=Ice; 7=Liquid Virga; 8=Aerosol
cloud_phase_mod = xr.where(cld_phase == 8, 0,  # Values of 8 become 0 - clear
                            xr.where((cld_phase == 3) | (cld_phase == 7), 1,  # Values of 3 and 7 become 1 - liquid
                                     xr.where((cld_phase == 5) | (cld_phase == 6), 2,  # Values of 5 and 6 become 2 - ice
                                              xr.where(cld_phase == 4, 3, cld_phase)  # Values of 4 become 3 - mixed
                                              )))

# Floor 'hour' values to the nearest 10
cloud_phase_mod['hour_floor'] = np.floor(cloud_phase_mod['hour'] / 10) * 10

cloud_phase_mod
```

```python
# Create new xarray dataset to store the averaged variables
lidar_radar_merged_um_reg_resolution = xr.Dataset()

# Select the variables
backscatter = lidar_radar_merged['merged_backscatter']
depolarization_ratio = lidar_radar_merged['depolarization_ratio']
reflectivity = lidar_radar_merged['reflectivity']
# var = var.fillna(0)
# var = xr.where(np.isinf(var), 0, var)

# Floor 'hour' values to the nearest 10
backscatter['hour_floor'] = np.floor(backscatter['hour'] / 10) * 10
depolarization_ratio['hour_floor'] = np.floor(depolarization_ratio['hour'] / 10) * 10
reflectivity['hour_floor'] = np.floor(reflectivity['hour'] / 10) * 10

# # Group by 'hour_floor' dimension and take the mean along 'height' dimension
averaged_backscatter = backscatter.groupby(backscatter['hour_floor']).mean(dim='hour')
averaged_depolarization_ratio = depolarization_ratio.groupby(depolarization_ratio['hour_floor']).mean(dim='hour')
averaged_reflectivity = reflectivity.groupby(reflectivity['hour_floor']).mean(dim='hour')

# Interpolate data from 'height' coordinate onto 'um_reg_heights' coordinate
averaged_backscatter = averaged_backscatter.interp(height=um_reg_heights)
averaged_depolarization_ratio = averaged_depolarization_ratio.interp(height=um_reg_heights)
averaged_reflectivity = averaged_reflectivity.interp(height=um_reg_heights)

# Assign the averaged variables to the new dataset
lidar_radar_merged_um_reg_resolution['merged_backscatter'] = averaged_backscatter
lidar_radar_merged_um_reg_resolution['depolarization_ratio'] = averaged_depolarization_ratio
lidar_radar_merged_um_reg_resolution['reflectivity'] = averaged_reflectivity

# Rename coordinates and remove data above 15 km
lidar_radar_merged_um_reg_resolution = lidar_radar_merged_um_reg_resolution.rename({'height': 'um_reg_heights', 'hour_floor': 'um_reg_minutes'})
lidar_radar_merged_um_reg_resolution = lidar_radar_merged_um_reg_resolution.where(lidar_radar_merged_um_reg_resolution.um_reg_heights < 15000, drop=True)

lidar_radar_merged_um_reg_resolution
```

```python
# Set the height limit for the plots
ylim_max = 15000

# Create a meshgrid for time and height to be used for all plots produced from the current file
time_mesh, height_mesh = np.meshgrid(lidar_radar_merged_um_reg_resolution['um_reg_minutes'], lidar_radar_merged_um_reg_resolution['um_reg_heights'])

# Backscatter
# Extract the desired variable
merged_backscatter = lidar_radar_merged_um_reg_resolution['merged_backscatter']
# Create the plot with a grey background
plt.figure(figsize = (12, 6))
plt.gca().set_facecolor('lightgrey')
# Create and organise a pcolormesh plot
pcm = plt.pcolormesh(time_mesh, height_mesh, merged_backscatter, cmap = 'viridis', shading = 'auto')
plt.xlabel('Time')
plt.ylabel('Height (m)')
plt.colorbar(pcm, label = 'Backscatter (log10 /m /sr)')
plt.title('Backscatter (averaged) vs. time and height')
plt.ylim(0,ylim_max)
plt.xticks(rotation = 45)
plt.tight_layout()
# Save and show the plot
averaged_backscatter = plt.gcf()
averaged_backscatter.savefig(f'/g/data/jk72/ck4840/projects/emc2/figs/marcus/merged_lidar_radar/um_reg_resolution/marcus_backscatter_{date_str}.png', dpi=200)
plt.show()

# Depolarization ratio
# Extract the desired variable
depolarization_ratio = lidar_radar_merged_um_reg_resolution['depolarization_ratio']
# Create the depolarization ratio plot
plt.figure(figsize=(12, 6))
# Set the entire plot background color to grey
plt.gca().set_facecolor('lightgrey')
# Create a pcolormesh plot
pcm = plt.pcolormesh(time_mesh, height_mesh, depolarization_ratio, cmap='viridis', vmin = 0, vmax = 1)
# Set plot labels, limits, and layout
plt.colorbar(pcm, label = 'Depolarization ratio')
plt.xlabel('Time')
plt.ylabel('Range')
plt.ylim(0,ylim_max)
plt.title('Depolarization ratio vs. time and altitude')
plt.xticks(rotation=45)
plt.tight_layout()
# Save and show the plot
averaged_depolarization_ratio = plt.gcf()
averaged_depolarization_ratio.savefig(f'/g/data/jk72/ck4840/projects/emc2/figs/marcus/merged_lidar_radar/um_reg_resolution/marcus_depolarization_ratio_{date_str}.png', dpi=200)
plt.show()

# Reflectivity
# Extract the desired variable
reflectivity = lidar_radar_merged_um_reg_resolution['reflectivity']
# Create the reflectivity plot
plt.figure(figsize=(12, 6))
# Set the entire plot background color to grey
plt.gca().set_facecolor('lightgrey')
# Create a pcolormesh plot
pcm = plt.pcolormesh(time_mesh, height_mesh, reflectivity, cmap='viridis')
# Set plot labels, limits, and layout
plt.colorbar(pcm, label='Reflectivity (dBZ)')
plt.xlabel('Time')
plt.ylabel('Range')
plt.ylim(0,ylim_max)
plt.title('Reflectivity vs. time and altitude')
plt.xticks(rotation=45)
plt.tight_layout()
# Save and show the plot
averaged_reflectivity = plt.gcf()
averaged_reflectivity.savefig(f'/g/data/jk72/ck4840/projects/emc2/figs/marcus/merged_lidar_radar/um_reg_resolution/marcus_reflectivity_{date_str}.png', dpi=200)
plt.show()
```
