Create the list of model output files to load, subset, and clean up

```python
# Initialise the empty list
ACCESS_AM2_list = []
# Walk through the directory tree and search for 'met' in filenames
for root, dirs, files in os.walk(os.getcwd()):
    for filename in files:
        # Check if the filename contains 'met'
        if 'met' in filename:
            # If 'met' is present, create the full file path and append it to the list
            file_path = os.path.join(root, filename)
            ACCESS_AM2_list.append(file_path)

# Arrange the list by file date
# Define a function to extract the date from the file path
def extract_date(path):
    # Split the path by '/'
    path_split = path.split('.')
    # Get the part containing the date (e.g. '20170901')
    date_part = next(part for part in path_split if part.startswith('met20'))
    # Extract and return the date portion (e.g., '20170901')
    return date_part[3:11]

# Define a function to extract the numeric part of the filename
def extract_numeric_filename(filename):
    return int(filename.split('_')[-1].split('.')[2][3:11]) # Extract the numeric part and convert to integer

# Sort ACCESS_AM2_list based on the extracted date and numeric part of the filename
ACCESS_AM2_list = sorted(
    ACCESS_AM2_list,
    key = lambda path: (extract_date(path), extract_numeric_filename(os.path.basename(path))))

# Define the cutoff dates
cutoff_dates = [datetime(2017, 10, 29), datetime(2018, 2, 28)]

# Create a new list to store filtered files
ACCESS_AM2_list_filtered = []

# Iterate over each file in ACCESS_AM2_list
for file in ACCESS_AM2_list:
    # Extract the date string from the file path
    date_str = datetime.strptime(file.split('/')[-1].split('.')[2][-8:], '%Y%m%d')
    
    # Check if the date is within the specified bounds
    if cutoff_dates[0] <= date_str <= cutoff_dates[-1]:
        # If the date is within the bounds, add the file to the filtered list
        ACCESS_AM2_list_filtered.append(file)

# Create the list of unique files
ACCESS_AM2_list_unique = np.unique(ACCESS_AM2_list_filtered)
# ACCESS_AM2_list_unique
```

```python
# Define the directory paths
bx400_directory = '/g/data/jk72/slf563/ACCESS/output/bx400/daily/'
# Directory options for field2450
# ./dd154/daily/dd154a.pdch
# ./cq686/daily/cq686a.pdch
# ./cq687/daily/cq687a.pdch
# ./ch543/daily/ch543a.pdch
dd154_directory = '/g/data/jk72/slf563/ACCESS/output/dd154/daily/'

# Define the patterns
bx400_pattern = bx400_directory + 'bx400a.pdc2*'
dd154_pattern = dd154_directory + 'dd154a.pdch2*'

# Use glob to find all files matching the patterns
bx400_list = glob.glob(bx400_pattern)
dd154_list = glob.glob(dd154_pattern)

# Arrange the lists by file date
# Define functions to extract the date from the file paths
def extract_date_bx400(path):
    # Split the path by '/'
    path_split = path.split('/')
    # Get the part containing the date (e.g. '20170901')
    date_part = next(part for part in path_split if part.startswith('bx400a.pdc2'))
    # Extract and return the date portion (e.g., '20170901')
    return date_part[10:18]
def extract_date_dd154(path):
    # Split the path by '/'
    path_split = path.split('/')
    # Get the part containing the date (e.g. '20170901')
    date_part = next(part for part in path_split if part.startswith('dd154a.pdch2'))
    # Extract and return the date portion (e.g., '20170901')
    return date_part[10:18]

# Define functions to extract the numeric part of the filenames
def extract_numeric_filename_bx400(filename):
    return int(filename.split('.')[1][3:11]) # Extract the numeric part and convert to integer
def extract_numeric_filename_dd154(filename):
    return int(filename.split('.')[1][4:12]) # Extract the numeric part and convert to integer

# Sort lists based on the extracted date and numeric part of the filenames
bx400_list = sorted(
    bx400_list,
    key = lambda path: (extract_date_bx400(path), extract_numeric_filename_bx400(os.path.basename(path))))
dd154_list = sorted(
    dd154_list,
    key = lambda path: (extract_date_dd154(path), extract_numeric_filename_dd154(os.path.basename(path))))

# Define the cutoff dates
cutoff_dates = [datetime(2017, 10, 29), datetime(2018, 2, 28)]

# Create new lists to store filtered files
bx400_list_sorted = []
dd154_list_sorted = []

# Iterate over each file in bx400_list
for file in bx400_list:
    # Extract the date string from the file path
    date_str = datetime.strptime(file.split('.')[1][3:11], '%Y%m%d')
    
    # Check if the date is within the specified bounds
    if cutoff_dates[0] <= date_str <= cutoff_dates[-1]:
        # If the date is within the bounds, add the file to the filtered list
        bx400_list_sorted.append(file)
# Iterate over each file in dd154_list
for file in dd154_list:
    # Extract the date string from the file path
    date_str = datetime.strptime(file.split('.')[1][4:12], '%Y%m%d')
    
    # Check if the date is within the specified bounds
    if cutoff_dates[0] <= date_str <= cutoff_dates[-1]:
        # If the date is within the bounds, add the file to the filtered list
        dd154_list_sorted.append(file)

# Create the list of unique files
bx400_list_unique = np.unique(bx400_list_sorted)
dd154_list_unique = np.unique(dd154_list_sorted)
# bx400_list_unique
# dd154_list_unique
```
