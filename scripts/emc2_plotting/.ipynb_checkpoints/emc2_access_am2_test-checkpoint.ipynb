Create the list of model output files to load, subset, and clean up

```python
# Initialise the empty list
ACCESS_AM2_list = []
# Walk through the directory tree and search for 'met' in filenames
for root, dirs, files in os.walk(os.getcwd()):
    for filename in files:
        # Check if the filename contains 'met'
        if 'met' in filename:
            # If 'met' is present, create the full file path and append it to the list
            file_path = os.path.join(root, filename)
            ACCESS_AM2_list.append(file_path)

# Arrange the list by file date
# Define a function to extract the date from the file path
def extract_date(path):
    # Split the path by '/'
    path_split = path.split('.')
    # Get the part containing the date (e.g. '20170901')
    date_part = next(part for part in path_split if part.startswith('met20'))
    # Extract and return the date portion (e.g., '20170901')
    return date_part[3:11]

# Define a function to extract the numeric part of the filename
def extract_numeric_filename(filename):
    return int(filename.split('_')[-1].split('.')[2][3:11]) # Extract the numeric part and convert to integer

# Sort UM_reg_list based on the extracted date and numeric part of the filename
ACCESS_AM2_list = sorted(
    ACCESS_AM2_list,
    key = lambda path: (extract_date(path), extract_numeric_filename(os.path.basename(path)))
)

# Display the sorted list
print(len(ACCESS_AM2_list))
ACCESS_AM2_list[157]
```

Load, subset, clean, and resave ACCESS-AM2 model data

```python
# Create the list of unique ACCESS-AM2 daily files between 2017-10-29 and 2018-02-28
ACCESS_AM2_list_unique = np.unique(ACCESS_AM2_list)
# TEST 1 DAY
ACCESS_AM2_list_unique = [ACCESS_AM2_list_unique[157]]
print(ACCESS_AM2_list_unique)

# Loop through ACCESS_AM2_list_unique
for file in ACCESS_AM2_list_unique:
    ACCESS_AM2_output = xr.open_dataset(file)

    # Drop unnecessary variables from the dataset
    # Drop the coordinates lat_v, time1, and z0_hybrid_height
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('lat_v')
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('time1')
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('z0_hybrid_height')
    # Drop fields deemed unnecessary for EMC2
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('va') # V component of wind

    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('theta') # Potential temperature
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('bldepth') # Boundary layer depth
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('sic') # Sea ice fraction
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('sit') # Sea ice depth
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field150') # W component of wind
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field265') # Area cloud fraction
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field266') # Bulk cloud fraction
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field267') # Liquid cloud fraction
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field268') # Frozen cloud fraction
    # ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field272') # Rain number concentration, else mixing ratio if consistent with UM_reg
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field392') # Cloud liquid mixing ratio
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field393') # Cloud ice mixing ratio
    # ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field394') # Rain mixing ratio
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field395') # Graupel mixing ratio
    # ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field396') # Ice crystal (precip?) mixing ratio
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('ps') # Surface pressure
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('ts') # Sea surface temperature
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('ts_1') # Sea-ice surface temperature
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field509') # Sea ice albedo
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('rss') # Net down surface SW flux
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('rsdt') # Incoming TOA SW flux
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('rsut') # Outgoing TOA SW flux
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('rsutcs') # Clear sky outgoing TOA SW flux
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('rsdscs') # Clear sky incoming surface SW flux
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('rsuscs') # Clear sky outgoing surface SW flux
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('rsds') # Total incoming surface SW flux
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('rls') # Net incoming surface LW flux
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('clt') # Total cloud amount in LW radiation (cloud area fraction)
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('rlut') # Outgoing TOA LW flux
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('rlutcs') # Clear sky outgoing TOA LW flux
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('rlds') # Incoming surface LW flux
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('rldscs') # Clear sky incoming surface LW flux
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field2330') # COSP cloud weights
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field2451') # COSP cloud area
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field2452') # COSP liquid cloud fraction
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field2453') # COSP ice cloud fraction
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field2454') # COSP high-level cloud fraction
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field2455') # COSP mid-level cloud fraction
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field2456') # COSP low-level cloud fraction
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field2457') # COSP tau
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field2463') # COSP Reff liquid
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field2464') # COSP Reff ice
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field2465') # COSP cloud top pressure
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field2466') # COSP liquid water path
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field2467') # COSP ice water path
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('vas') # Northward wind
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('hfss') # Surface upward sensible heat flux
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field3230') # 10 metre wind speed on C grid
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('hfls') # Surface upward latent heat flux
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('tas') # Temperature at 1.5 m
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('huss') # Specific humidity at 1.5 m
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field3245') # Relative humidity at 1.5 m
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('rlut_1') # Outgoing TOA LW flux after BL
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field4113') # Graupel radar reflectivity
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field4114') # Ice aggregate radar reflectivity
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field4115') # Ice crystal radar reflectivity
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field4116') # Rain radar reflectivity
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field4117') # Liquid cloud radar reflectivity
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field4203') # Large scale rainfall rate
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field4204') # Large scale snowfall rate
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field5205') # Convective rainfall rate
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field5206') # Convective snowfall rate
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('prrn') # Total rainfall rate
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('prsn') # Total snowfall rate
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('pr') # Total precipitation rate
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field9203') # Low cloud amount
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field9204') # Medium cloud amount
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field9205') # High cloud amount
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('psl') # Air pressure at sea level
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('lwp') # Cloud liquid water content
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('clivi') # Cloud ice content
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field30453') # Height at tropopause level
    ACCESS_AM2_output = ACCESS_AM2_output.drop_vars('field30461') # Total column Q    

    # Extract the date from the 'time' coordinate in the model data and filter the MARCUS coordinate data to the model data date
    marcus_coordinates_current = marcus_coordinates_v1_3[marcus_coordinates_v1_3['datetime_day'].dt.date == pd.to_datetime(ACCESS_AM2_output['time'].values).date[0]]
    # Convert MARCUS coordinate data to XArray dataset and convert datetime_hour variable to a coordinate
    marcus_coordinates_current = xr.Dataset.from_dataframe(marcus_coordinates_current)
    marcus_coordinates_current = marcus_coordinates_current.set_coords(['datetime_hour'])

    # Subset the ACCESS-AM2 model data to the MARCUS Underway coordinates
    # Extract the time values from the ACCESS-AM2 output
    time_values = ACCESS_AM2_output['time'].values
    time_dataset = xr.Dataset({'time': time_values})
    # Initialize lists to store latitudes and longitudes
    lat_values = []
    lon_values = []
    # Initalise list to store the individual datasets from each iteration
    list_of_datasets = []
    # Loop over each time step in the ship dataset
    for i in range(len(marcus_coordinates_current['datetime_hour'])):
        # Get the lat, lon, and time values for each step
        lat = marcus_coordinates_current['latitude'].isel(index=i)
        lon = marcus_coordinates_current['longitude'].isel(index=i)
        time = marcus_coordinates_current['datetime_hour'].isel(index=i)
        # Extract the current lat and lon values
        lat_values.append(lat)
        lon_values.append(lon)
        # Interpolating the entire dataset at the ship's position
        ACCESS_AM2_output_interpolated = ACCESS_AM2_output.interp(
            lat=lat.item(),
            lon=lon.item(),
            time=time_values[i],
            method='nearest')
        # Drop the current latitude and longitude coordinates
        ACCESS_AM2_output_interpolated = ACCESS_AM2_output_interpolated.drop_vars('lat')
        ACCESS_AM2_output_interpolated = ACCESS_AM2_output_interpolated.drop_vars('lon')
        # Append each interpolated dataset to the list
        list_of_datasets.append(ACCESS_AM2_output_interpolated)
    # Concatenate all datasets along a new dimension 'index'
    ACCESS_AM2_output_subset = xr.combine_nested(list_of_datasets, concat_dim='time')
    # Merge the 'time' dataset with the combined dataset
    ACCESS_AM2_output_subset = xr.merge([time_dataset, ACCESS_AM2_output_subset])
    # Add latitudes and longitudes as variables
    lat_values = xr.DataArray(data=lat_values, dims='dim_name')
    lon_values = xr.DataArray(data=lon_values, dims='dim_name')
    ACCESS_AM2_output_subset['lat'] = ('time', lat_values.data)
    ACCESS_AM2_output_subset['lon'] = ('time', lon_values.data)

    # Add a utility zero array to the dataset
    # Create a variable filled with zeros
    zeros_data = xr.DataArray(
        data = np.zeros((len(ACCESS_AM2_output_subset['time']), len(ACCESS_AM2_output_subset['z1_hybrid_height']))),
        dims = ('time', 'z1_hybrid_height'),
        coords = {'time': ACCESS_AM2_output_subset['time'], 'z1_hybrid_height': ACCESS_AM2_output_subset['z1_hybrid_height']})
    # Add the zeros variable to the dataset
    ACCESS_AM2_output_subset['zeros_var'] = zeros_data

    # Add a 2-dimensional z_values variable to the dataset
    # Extract the z1_hybrid_height coordinate from the dataset
    z1_values = ACCESS_AM2_output_subset['z1_hybrid_height'].values
    # Create 2D array z_values from 1D array z1_values
    z_values = z1_values[:, np.newaxis] * np.ones(len(ACCESS_AM2_output_subset['time']))
    # Reinsert z variable into the dataset
    ACCESS_AM2_output_subset['z'] = (('z1_hybrid_height', 'time'), z_values)
    ACCESS_AM2_output_subset['z'].attrs['units'] = 'meters'
    ACCESS_AM2_output_subset['z'] = ACCESS_AM2_output_subset['z'].transpose('time', 'z1_hybrid_height')

    # Estimate particle effective radius for each species and add it to the data subset
    # Cloud liquid
    # Extract/calculate required variables and set constants
    qcl = ACCESS_AM2_output_subset.field254 # kg/kg (unitless)
    rhoa = ACCESS_AM2_output_subset.field408/(ACCESS_AM2_output_subset.ta*287.058) # kg/m^3
    ncl = ACCESS_AM2_output_subset.field4210 # particles/m^3
    ncl_mass = ncl/rhoa # particles/kg
    # Parameterisation
    re_cl = ((((qcl/ncl_mass)/1000)*3)/(4*np.pi))**(1./3)
    # Cloud ice
    # Extract/calculate required variables and set constants
    qci = ACCESS_AM2_output_subset.field12 # kg/kg
    rhoa = ACCESS_AM2_output_subset.field408/(ACCESS_AM2_output_subset.ta*287.058) # kg/m^3
    ta = ACCESS_AM2_output_subset.ta
    # Parameterisation
    Zqci = qci * rhoa * 1000.
    ZRefDe = 0.64952
    Z_TEMPC = ta - 83.15
    Z_TCELS = ta - 273
    Z_FSR = 1.2351 + 0.0105 * Z_TCELS
    ZAqci = 45.8966 * (Zqci ** 0.2214)
    ZBqci = 0.7957 * (Zqci ** 0.2535)
    Z_DESR = Z_FSR * (ZAqci + ZBqci * Z_TEMPC)
    # Z_DESR = (30. > Z_TEMPC < 155.) - reference range - values should be within these micron bounds
    re_ci = ZRefDe * Z_DESR * 1e-6
    # Precipitating liquid
    # Note - there is no data for precipitating liquid particle number density, so currently a zeros array is used
    # NB: FIELD 272 IS RAIN AFTER TIMESTEP AND MAY BE RAIN NUMBER CONCENTRATION
    # Extract/calculate required variables and set constants
    qpl = ACCESS_AM2_output_subset.field394 # kg/kg (unitless)
    rhoa = ACCESS_AM2_output_subset.field408/(ACCESS_AM2_output_subset.ta*287.058) # kg/m^3
    npl = ACCESS_AM2_output_subset.zeros_var # particles/m^3
    npl_mass = npl/rhoa # particles/kg
    # Parameterisation
    re_pl = ((((qpl/npl_mass)/1000)*3)/(4*np.pi))**(1./3)
    # Precipitating ice
    # Extract/calculate required variables and set constants
    qpi = ACCESS_AM2_output_subset.field396 # kg/kg
    rhoa = ACCESS_AM2_output_subset.field408/(ACCESS_AM2_output_subset.ta*287.058) # kg/m^3
    ta = ACCESS_AM2_output_subset.ta
    # Parameterisation
    Zqpi = qpi * rhoa * 1000.
    ZRefDe = 0.64952
    Z_TEMPC = ta - 83.15
    Z_TCELS = ta - 273
    Z_FSR = 1.2351 + 0.0105 * Z_TCELS
    ZAqpi = 45.8966 * (Zqpi ** 0.2214)
    ZBqpi = 0.7957 * (Zqpi ** 0.2535)
    Z_DESR = Z_FSR * (ZAqpi + ZBqpi * Z_TEMPC)
    # Z_DESR = (30. > Z_TEMPC < 155.) - reference range - values should be within these micron bounds
    re_pi = ZRefDe * Z_DESR * 1e-6
    # Add effective radius to the dataset
    ACCESS_AM2_output_subset['re_cl'] = xr.DataArray(data = re_cl,dims = ('time','z1_hybrid_height'))
    ACCESS_AM2_output_subset['re_ci'] = xr.DataArray(data = re_ci,dims = ('time','z1_hybrid_height'))
    ACCESS_AM2_output_subset['re_pl'] = xr.DataArray(data = re_pl,dims = ('time','z1_hybrid_height'))
    ACCESS_AM2_output_subset['re_pi'] = xr.DataArray(data = re_pi,dims = ('time','z1_hybrid_height'))
    # Replace effective radius nans with zeros
    ACCESS_AM2_output_subset['re_cl'] = ACCESS_AM2_output_subset['re_cl'].fillna(0)
    ACCESS_AM2_output_subset['re_ci'] = ACCESS_AM2_output_subset['re_ci'].fillna(0)
    ACCESS_AM2_output_subset['re_pl'] = ACCESS_AM2_output_subset['re_pl'].fillna(0)
    ACCESS_AM2_output_subset['re_pi'] = ACCESS_AM2_output_subset['re_pi'].fillna(0)
    # Replace effective radius inf with zeros
    ACCESS_AM2_output_subset['re_cl'] = xr.where(np.isinf(ACCESS_AM2_output_subset['re_cl']), 0, ACCESS_AM2_output_subset['re_cl'])
    ACCESS_AM2_output_subset['re_ci'] = xr.where(np.isinf(ACCESS_AM2_output_subset['re_ci']), 0, ACCESS_AM2_output_subset['re_ci'])
    ACCESS_AM2_output_subset['re_pl'] = xr.where(np.isinf(ACCESS_AM2_output_subset['re_pl']), 0, ACCESS_AM2_output_subset['re_pl'])
    ACCESS_AM2_output_subset['re_pi'] = xr.where(np.isinf(ACCESS_AM2_output_subset['re_pi']), 0, ACCESS_AM2_output_subset['re_pi'])
    # Print effective radius maximum values for each species
    print(ACCESS_AM2_output_subset['re_cl'].max())
    print(ACCESS_AM2_output_subset['re_ci'].max())
    print(ACCESS_AM2_output_subset['re_pl'].max())
    print(ACCESS_AM2_output_subset['re_pi'].max())

    # Add units to variables in the data subset
    # Pressure
    ACCESS_AM2_output_subset['field408'] = ACCESS_AM2_output_subset['field408'].assign_attrs(units=ureg.pascal)
    ACCESS_AM2_output_subset['field408'].attrs['units'] = str(ACCESS_AM2_output_subset['field408'].attrs['units'])
    # Temperature
    ACCESS_AM2_output_subset['ta'] = ACCESS_AM2_output_subset['ta'].assign_attrs(units=ureg.kelvin)
    ACCESS_AM2_output_subset['ta'].attrs['units'] = str(ACCESS_AM2_output_subset['ta'].attrs['units'])
    # Effective radius of particles
    ACCESS_AM2_output_subset['re_cl'] = ACCESS_AM2_output_subset['re_cl'].assign_attrs(units=ureg.meter)
    ACCESS_AM2_output_subset['re_cl'].attrs['units'] = str(ACCESS_AM2_output_subset['re_cl'].attrs['units'])
    ACCESS_AM2_output_subset['re_ci'] = ACCESS_AM2_output_subset['re_ci'].assign_attrs(units=ureg.meter)
    ACCESS_AM2_output_subset['re_ci'].attrs['units'] = str(ACCESS_AM2_output_subset['re_ci'].attrs['units'])
    ACCESS_AM2_output_subset['re_pl'] = ACCESS_AM2_output_subset['re_pl'].assign_attrs(units=ureg.meter)
    ACCESS_AM2_output_subset['re_pl'].attrs['units'] = str(ACCESS_AM2_output_subset['re_pl'].attrs['units'])
    ACCESS_AM2_output_subset['re_pi'] = ACCESS_AM2_output_subset['re_pi'].assign_attrs(units=ureg.meter)
    ACCESS_AM2_output_subset['re_pi'].attrs['units'] = str(ACCESS_AM2_output_subset['re_pi'].attrs['units'])
    # Water vapor mixing ratio
    ACCESS_AM2_output_subset['hus'] = ACCESS_AM2_output_subset['hus'].assign_attrs(units=ureg.meter**3/ureg.meter**3)
    ACCESS_AM2_output_subset['hus'].attrs['units'] = str(ACCESS_AM2_output_subset['hus'].attrs['units'])
    # Cloud droplet number concentration
    ACCESS_AM2_output_subset['field4210'] = ACCESS_AM2_output_subset['field4210'].assign_attrs(units=ureg.meter**(-3))
    ACCESS_AM2_output_subset['field4210'].attrs['units'] = str(ACCESS_AM2_output_subset['field4210'].attrs['units'])
    # Height field
    ACCESS_AM2_output_subset['z1_hybrid_height'] = ACCESS_AM2_output_subset['z1_hybrid_height'].assign_attrs(units=ureg.meter)
    ACCESS_AM2_output_subset['z1_hybrid_height'].attrs['units'] = str(ACCESS_AM2_output_subset['z1_hybrid_height'].attrs['units'])
    # Mass mixing ratios
    ACCESS_AM2_output_subset['field254'] = ACCESS_AM2_output_subset['field254'].assign_attrs(units=ureg.kg/ureg.kg)
    ACCESS_AM2_output_subset['field254'].attrs['units'] = str(ACCESS_AM2_output_subset['field254'].attrs['units'])
    ACCESS_AM2_output_subset['field12'] = ACCESS_AM2_output_subset['field12'].assign_attrs(units=ureg.kg/ureg.kg)
    ACCESS_AM2_output_subset['field12'].attrs['units'] = str(ACCESS_AM2_output_subset['field12'].attrs['units'])
    ACCESS_AM2_output_subset['field394'] = ACCESS_AM2_output_subset['field394'].assign_attrs(units=ureg.kg/ureg.kg)
    ACCESS_AM2_output_subset['field394'].attrs['units'] = str(ACCESS_AM2_output_subset['field394'].attrs['units'])
    ACCESS_AM2_output_subset['field396'] = ACCESS_AM2_output_subset['field396'].assign_attrs(units=ureg.kg/ureg.kg)
    ACCESS_AM2_output_subset['field396'].attrs['units'] = str(ACCESS_AM2_output_subset['field396'].attrs['units'])

# marcus_coordinates_current
ACCESS_AM2_output_subset
```

Create the ACCESS-AM2 model class

```python
class ACCESS_AM2(emc2.core.Model):
    def __init__(self, file_path):
       """
       This loads an ACCESS-AM2 simulation with all of the necessary parameters for EMC^2 to run.
       
       Parameters
       ----------
       file_path: str
           Path to an ACCESS-AM2 simulation.
       time_range: tuple, list, or array, typically in datetime64 format
           Two-element array with starting and ending of time range.
       load_processed: bool
           If True, treating the 'file_path' variable as an EMC2-processed dataset; thus skipping dimension stacking as part of pre-processing.
       """
       
       super().__init__()
       
       # Bulk density
       self.Rho_hyd = {'cl': 1000. * ureg.kg / (ureg.m**3),
                       'ci': 500.  * ureg.kg / (ureg.m**3),
                       'pl': 1000. * ureg.kg / (ureg.m**3),
                       'pi': 250.  * ureg.kg / (ureg.m**3)}
       self.fluffy = {'ci': 0.5 * ureg.dimensionless,
                      'pi': 0.5 * ureg.dimensionless}
       # Lidar ratio
       self.lidar_ratio = {'cl': 18.  * ureg.dimensionless,
                           'ci': 24.  * ureg.dimensionless,
                           'pl': 5.5  * ureg.dimensionless,
                           'pi': 24.0 * ureg.dimensionless}
       # Lidar LDR per hydrometeor mass content
       self.LDR_per_hyd = {'cl': 0.03 * 1 / (ureg.kg / (ureg.m**3)),
                           'ci': 0.35 * 1 / (ureg.kg / (ureg.m**3)),
                           'pl': 0.1  * 1 / (ureg.kg / (ureg.m**3)),
                           'pi': 0.40 * 1 / (ureg.kg / (ureg.m**3))}
       # a, b in V = aD^b
       self.vel_param_a = {'cl': 3e7,
                           'ci': 700.,
                           'pl': 841.997,
                           'pi': 11.72}
       self.vel_param_b = {'cl': 2.   * ureg.dimensionless,
                           'ci': 1.   * ureg.dimensionless,
                           'pl': 0.8  * ureg.dimensionless,
                           'pi': 0.41 * ureg.dimensionless}
       super()._add_vel_units()
       # Names of mixing ratios of species
       # What is the difference between field254 and field392? Also, what is the difference between field12 and field393?
       self.q_names = {'cl': 'field254', 'ci': 'field12', 'pl': 'field394', 'pi': 'field396'}
       # Number concentration of each species
       self.N_field = {'cl': 'field4210', 'ci': 'zeros_var', 'pl': 'zeros_var', 'pi': 'zeros_var'} # Need number concentrations of each species
       # Convective fraction
       self.conv_frac_names = {'cl': 'zeros_var', 'ci': 'zeros_var', 'pl': 'zeros_var', 'pi': 'zeros_var'}
       self.conv_frac_names_for_rad = {'cl': 'zeros_var', 'ci': 'zeros_var', 'pl': 'zeros_var', 'pi': 'zeros_var'}
       # Stratiform fraction
       self.strat_frac_names = {'cl': 'zeros_var', 'ci': 'zeros_var', 'pl': 'zeros_var', 'pi': 'zeros_var'}
       self.strat_frac_names_for_rad = {'cl': 'zeros_var', 'ci': 'zeros_var', 'pl': 'zeros_var', 'pi': 'zeros_var'}
       # Effective radius
       self.re_fields = {'cl': 're_cl', 'ci': 're_ci', 'pl': 're_pl', 'pi': 're_pi'}
       self.strat_re_fields = {'cl': 're_cl', 'ci': 're_ci', 'pl': 're_pl', 'pi': 're_pi'}
       self.conv_re_fields = {'cl': 'zeros_var', 'ci': 'zeros_var', 'pl': 'zeros_var', 'pi': 'zeros_var'}
       # Convective mixing ratio
       self.q_names_convective = {'cl': 'zeros_var', 'ci': 'zeros_var', 'pl': 'zeros_var', 'pi': 'zeros_var'}
       # Stratiform mixing ratio
       self.q_names_stratiform = {'cl': 'field254', 'ci': 'field12', 'pl': 'field394', 'pi': 'field396'}
       # self.q_names_stratiform = {'cl': 'zeros_var', 'ci': 'zeros_var', 'pl': 'zeros_var', 'pi': 'zeros_var'}
       # Water vapor mixing ratio
       self.q_field = "hus"
       # Pressure
       self.p_field = "field408"
       # Height
       self.z_field = "z"
       # Temperature
       self.T_field = "ta"
       # Name of height dimension
       self.height_dim = "z1_hybrid_height"
       # Name of time dimension
       self.time_dim = "time"
       self.hyd_types = ["cl", "ci", "pl", "pi"]
       self.process_conv = False
       self.model_name = "ACCESS_AM2"
       # self.ds = xr.open_dataset(file_path)
       self.ds = met20180219_subset
```
